# Tracking the Sun

```python
from pyathena.connection import Connection
from pyathena.pandas_cursor import PandasCursor
```

```python
cursor = Connection(
    profile_name="dev",
    region_name="us-west-2",
    s3_staging_dir="s3://nrel-tests/tracking_the_sun/"
).cursor(PandasCursor)
```

## 1. Map

```python
import folium
from bokeh.sampledata.us_states import data as states
from bokeh.plotting import figure, show
```

```python
df1 = cursor.execute(
    """
    SELECT state, date_trunc('installtion date') AS installation_year, COUNT() AS count
    FROM oedidb.oedi_tracking_the_sun
    GROUP BY installation_year
    """
).as_pandas()
```

```python
df1.head()
```

```python
p = figure(plot_width=800, plot_height=400)
p.vbar(x=range(len(df1.state)), width=0.5, bottom=0, top=df1["count"].astype("int").values(), color="firebrick")
show(p)
```
